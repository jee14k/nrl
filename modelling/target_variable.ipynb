```python
import pandas as pd 
```

```python
broncos = pd.read_csv('../data_preprocessing/match_level_data/broncos_games_with_team_stats.csv')
```

```python
broncos.head()
```

```python
def get_team_result(row):
    try:
        # Extract scores
        home_score, away_score = [int(x.strip()) for x in row['Result'].split('-')]
        team_name = str(row['Team']).lower()
        home_team = str(row['Home Team']).lower()
        away_team = str(row['Away Team']).lower()

        # Match by checking if any word in home/away team appears in Team
        if any(word in team_name for word in home_team.split()):
            return 'Win' if home_score > away_score else 'Loss'
        elif any(word in team_name for word in away_team.split()):
            return 'Win' if away_score > home_score else 'Loss'
        else:
            return None  # Team name doesn’t match home or away
    except:
        return None
```

```python
broncos['Result_Label'] = broncos.apply(get_team_result, axis=1)
```
