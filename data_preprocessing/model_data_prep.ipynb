```python
import pandas as pd 

season_stats_df = pd.read_csv("../data/team_stats/nrl_2025_team_stats_cleaned.csv")
broncos_results = pd.read_csv("../data/previous_games/nrl-2025-broncos-UTC.csv")
```

```python
season_stats_df.head()
```

```python
broncos_results.head()
```

```python
broncos_results['Team'] = 'Brisbane Broncos'
```

```python
broncos_results.head()
```

```python
broncos_stats = season_stats_df[season_stats_df['team'] == 'Brisbane Broncos'].copy()
broncos_stats.columns = [f'team_{col}' if col != 'team' else 'Team' for col in broncos_stats.columns]
```

```python
broncos_stats.head()
```

```python
# Merge to all games
broncos_games = broncos_results.merge(broncos_stats, on='Team', how='left')
```

```python
broncos_games.head()
```

```python
def find_opponent(row):
    home_words = set(row['Home Team'].lower().split())
    team_words = set(row['Team'].lower().split())
    # Check if any word in home team is in team
    if home_words.intersection(team_words):
        return row['Away Team']
    else:
        return row['Home Team']

broncos_games['Opponent'] = broncos_games.apply(find_opponent, axis=1)

# Now keep only the last word of opponent names
broncos_games['Opponent'] = broncos_games['Opponent'].apply(lambda x: x.split()[-1])
```

```python
broncos_games.head()
```

```python
team_name_map = {
    'Rabbitohs': 'South Sydney Rabbitohs',
    'Cowboys': 'North Queensland Cowboys',
    'Panthers': 'Penrith Panthers',
    'Storm': 'Melbourne Storm',
    'Warriors': 'Warriors',
    'Titans': 'Gold Coast Titans',
    'Sharks': 'Cronulla-Sutherland Sharks',
    'Dolphins': 'The Dolphins',
    'Eagles': 'Manly-Warringah Sea Eagles',
    'Bulldogs': 'Canterbury-Bankstown Bulldogs',
    'Dragons': 'St George Illawarra Dragons',
    'Roosters': 'Sydney Roosters',
    'Tigers': 'Wests Tigers',
    'Eels': 'Parramatta Eels',
    'Raiders': 'Canberra Raiders',
    'Knights': 'Newcastle Knights',
    'Broncos': 'Brisbane Broncos'  # Optional for completeness
}
```

```python
broncos_games['Opponent'] = broncos_games['Opponent'].map(team_name_map)
```

```python
broncos_games.head()
```

```python
# Copy season stats for away teams and rename columns with 'away_' prefix
away_stats = season_stats_df.copy()
away_stats.columns = ['Opponent'] + [f'away_{col}' for col in season_stats_df.columns if col != 'team']
```

```python
away_stats.head()
```

```python
#Merge away stats into broncos_games using updated Opponent column
broncos_games = broncos_games.merge(away_stats, on='Opponent', how='left')
```

```python
print(broncos_games['Opponent'].unique())
print(away_stats['Opponent'].unique())
```

```python
import os

# Create the folder if it doesn't exist
folder_path = 'match_level_data'
os.makedirs(folder_path, exist_ok=True)

# Define the full file path
file_path = os.path.join(folder_path, 'broncos_games_with_team_stats.csv')

# Save the DataFrame to CSV
broncos_games.to_csv(file_path, index=False)
```
